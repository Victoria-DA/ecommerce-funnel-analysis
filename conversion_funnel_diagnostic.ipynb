```python
import plotly.graph_objects as go

# 1. Defining the Funnel Data (in English)
stages = ["Website Visitors", "Added to Cart", "Started Checkout", "Payment Window", "Completed Order"]
values = [30000, 19589, 14694, 12586, 0]

# 2. Creating the Funnel Chart
fig = go.Figure(go.Funnel(
    y = stages,
    x = values,
    textinfo = "value+percent initial",
    marker = {"color": ["#636EFA", "#EF553B", "#00CC96", "#AB63FA", "#FFA15A"]}
))

# 3. Customizing the layout
fig.update_layout(
    title_text="Sales Funnel Analysis: Online Food Orders",
    title_x=0.5
)

fig.show()
```

## 📝 Executive Summary
The funnel reveals a healthy conversion rate in the initial stages (Visitors to Payment Window). However, there is a **0% completion rate** at the final step.

## 🧐 Technical Hypotheses
As a Data Analyst, my diagnostic points to:
1. **Broken 'Complete Order' Button**: A potential JavaScript error preventing the form submission.
2. **Payment Integration Failure**: The API responsible for processing payments might be offline.
3. **UI/UX Blocker**: The final call-to-action (CTA) might not be visible on mobile devices.

## 🚀 Recommended Action
Assign a QA (Quality Assurance) engineer to test the checkout flow immediately.
